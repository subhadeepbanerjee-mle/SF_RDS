```python
#заполняем недостающие значения по количеству

df['bc_passengers'] = df['bc_passengers'].fillna(10.51)
df['ec_passengers'] = df['ec_passengers'].fillna(102.61)
df['fact_passengers'] = df.bc_passengers + df.ec_passengers

#заполняем недостающие значения по стоимости билетов
df['bc_reveneu'] = df['bc_reveneu'].fillna(3.846102e+05)
df['ec_reveneu'] = df['ec_reveneu'].fillna(1.257966e+06)
df['total_amount'] = df.ec_reveneu + df.ec_reveneu

# Заполняем недостающие значения относительной загруженности мест средними значениями

df['fraction'] = df['fraction'].map(lambda x: 8.700000e-01 if x == 0 else x)
```

```python
#данные по стоимости топлива в г. Анапа: (RUB/T)
# Источник <http://superjet.wikidot.com/wiki:rashod-topliva#toc1>
#Федеральное агеннство воздушного транспорта

f_cost_jan = 38867
f_cost_feb = 41435
f_cost_dec = 47101
```

```python
# конвертация данных в datetime
df['departure'] = pd.to_datetime(df['departure'])
```

```python
# для упрощения вычислений созадим временный столбец с номером месяца
df['departure_month'] = df['departure'].dt.month
```

```python
# Внесем данные в таблицу
df['fuel_cost'] = df['departure_month'].map(lambda x: f_cost_jan if x == 1  else (f_cost_feb if x == 2 else f_cost_dec))
```

```python
# удаляем временный столбец
df = df.drop(['departure_month'], axis=1)
```

```python
#расходы топлива моделей самолетов Т/Ч


B_737= 2600/1000 #Boeing 737-300
S_J = 1670/1000 #Sukhoi Superjet-100
```

```python
#внесем данные по расходу моделей в таблицу
df['fuel_consumption'] = df['model'].map(lambda x: B_737 if x == "Boeing 737-300"  else S_J)
```

```python
#вычисляем стоимость перелета по каждому рейсу
df['flight_cost'] = round(df.fuel_cost * df.fuel_consumption * df.duration,0)
```

```python
#вычисляем прибыльность рейса как разницу между выручкой и стоимостью топлива
df['flight_marge'] = df.total_amount - df.fuel_consumption
df
```

```python
#сводная таблица со средними значениями
res = df.groupby(['flight_no','arrival_city', 'model'])['duration', 'bc_passengers','ec_passengers','fraction','bc_reveneu','ec_reveneu','flight_marge'].mean().sort_values(by= 'flight_marge', ascending=True)
round(res,2)
```
