```python
url = 'https://auto.ru/-/ajax/desktop/listing/'


print('Start of parsing', datetime.datetime.now())

for Mark in models_lst[23:]:
    time.sleep(3)
    data = []

    for num in range(1,100):
        print(f'the page {num} of Mark {Mark} is being collected\n')

        params = {"category":"cars",
          "section":"used",
          "year_from":1990,
          "year_to":2021,
          "page":num,
          "catalog_filter": [{"mark": Mark}],
          "geo_radius":200,
          "geo_id":[213]
          }

        try:
            response = requests.post(url, json = params, headers = headers)
            response.raise_for_status()
        except requests.exceptions.HTTPError as e:
            print (e.response.text)
            continue

        page = response.json()['offers']

#FEATURES
        for card in page:
            # Root dir
            try: bodyType = card['vehicle_info']['configuration']['body_type']
            except: bodyType = ''

            try: brand = card['vehicle_info']['mark_info']['code']
            except: brand = ''

            try: color = card['vehicle_info']['complectation']['vendor_colors'][0]['stock_color']['name_ru']
            except: color = ''

            try: color_hex = card['color_hex']
            except: color_hex = 'NAN'

            try: complectation_dict = card['vehicle_info']['complectation']
            except: complectation_dict = ''

            try: description = card['description']
            except:  description = ''

            try: engineDisplacement = card['vehicle_info']['tech_param']['displacement']
            except: engineDisplacement = ''

            try:  enginePower = card['vehicle_info']['tech_param']['power']
            except: enginePower = ''

            try: equipment_dict = card['vehicle_info']['equipment']
            except: equipment_dict = ''

            try: fuelType = card['vehicle_info']['tech_param']['engine_type']
            except: fuelType = ''

            try: image = card['state']['image_urls'][0]['sizes']['1200x900']
            except: image = ''

            try:  mileage = card['state']['mileage']
            except: mileage= ''

            try: modelDate = card['vehicle_info']['super_gen']['year_from']
            except: modelDate = ''

            try: model_info =  card['vehicle_info']['model_info']
            except: model_info = ''

            try: model_name =  card['vehicle_info']['model_info']['name']
            except: model_name = ''

            try: name = card['vehicle_info']['tech_param']['human_name']
            except: name = ''

            try: numberOfDoors = card['vehicle_info']['configuration']['doors_count']
            except: numberOfDoors = 'NAN'

            parsing_unixtime = int(time.time())

            try: price = card['price_info']['price']
            except: price = ''

            try: priceCurrency = card['price_info']['currency']
            except: priceCurrency = ''

            try: productionDate = card['documents']['year']
            except: productionDate = ''

            try: sell_id = card['id']
            except: sell_id = ''

            try: super_gen = card['vehicle_info']['super_gen']
            except: super_gen = ''

            try: vehicleConfiguration = card['lk_summary']
            except: vehicleConfiguration = ''

            try: vehicleTransmission = card['vehicle_info']['tech_param']['transmission']
            except: vehicleTransmission = ''

            try: vendor = card['vehicle_info']['vendor']
            except: vendor = ''

            try: owners_number = card['documents']['owners_number']
            except: owners_number = ''

            try: holding_period_month = card['owner_expenses']['transport_tax']['holding_period_month']
            except: holding_period_month  = ''

            try: PTS = card['documents']['pts_original']
            except: PTS = ''

            try: gear_type = card['vehicle_info']['tech_param']['gear_type']
            except: gear_type = ''

            try: steering_wheel = card['vehicle_info']['steering_wheel']
            except: steering_wheel = ''

            try: state_not_beaten = card['state']['state_not_beaten']
            except: state_not_beaten = ''

            try: custom_cleared = card['documents']['custom_cleared']
            except: custom_cleared = ''


            data.append([
                bodyType, brand, color, color_hex, complectation_dict,
                description, engineDisplacement, enginePower, equipment_dict,
                fuelType, image, mileage, modelDate, model_info, model_name,
                name, numberOfDoors, parsing_unixtime, price, priceCurrency,
                productionDate, sell_id, super_gen, vehicleConfiguration,
                vehicleTransmission, vendor, owners_number, holding_period_month, PTS,
                gear_type, steering_wheel, state_not_beaten, custom_cleared
                ])

        time.sleep(0.5)

    fielname = Mark +'.json'
    target_path = './data/'

    with open(os.path.join(target_path, fielname), 'w') as json_file:
        json.dump(data, json_file)
        json_file.close()

print('the parsing was done successfully', datetime.datetime.now())
```
